```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import requests
import pandas as pd
from lxml import html
from selenium.webdriver.common.by import By

chrome_options = webdriver.ChromeOptions()
# 브라우저에 임의로 User-agent 옵션을 넣어 Python 코드로 접속함을 숨김
chrome_options.add_argument('--user-agent="Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/66.0.3359.139 Safari/537.36"')
driver = webdriver.Chrome(chrome_options=chrome_options)
```

```python
# 페이지 여러개 수집
# 추후에 vote 10개 초과만 택하자

# ~2022/01/01
article = []
for i in range(1,347) :

    driver.get("https://www1.president.go.kr/petitions/?c=0&only=2&page=" + str(i))
    
    soup = BeautifulSoup(driver.page_source, 'html.parser')

    for i in soup.select("#cont_view > div.cs_area > div > div > div.ct_petitions > div.ct_petitions_area.ct_txt_left > div.ct_list1 > div > div.b_list.category.b_list2 > div.bl_body > ul > li"):
        category = i.find("div", class_="bl_category ccategory cs wv_category").text[3:].strip()
        title = i.find("div", class_="bl_subject").text[3:].strip()
        end = i.find("div", class_="bl_date light").text[7:].strip()
        votes = i.find("div", class_="bl_agree cs").text[5:].strip()
        
        article_dict = {"category":category,"title":title,"end":end, "votes":votes}
        article.append(article_dict)

        
    time.sleep(2)
    
    
driver.close()
```

```python
data = pd.DataFrame(article)
data
```
